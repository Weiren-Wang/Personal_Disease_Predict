    测试数据处理

```python
test_df = pd.read_csv('test_df.csv',index_col = 0)
```

```python
test_df.columns = ['AAC001','true']
```

```python
def get_list_disease_from_readed_str(x):
    return x.replace('[','').replace(']','').replace("'",'').split(',')
```

```python
test_df['true'] = test_df.true.apply(lambda x:get_list_disease_from_readed_str(x))
```

```python
total_df = pd.merge(pred,test_df,on='AAC001')
```

    预测结果评估

```python
def calculate_predict_accuracy(df):
    duplicate_samples = 0
    true_samples = 0
    true_index = []
    false_samples = 0
    false_index = []
    for i,row in df.iterrows():
        if row.true[0] in row.input:
            duplicate_samples += 1
            continue
        else:
            flag = False
            for disease in row.pred:
                if disease in row.true:
                    flag = True
                    break
            if flag:
                true_samples += 1
                true_index.append(i)
            else:
                false_samples += 1
                false_index.append(i)
    print('############################### TEST REPORT ###################################')
    print('Total test samples : ',len(df))
    print('Valid test samples : ',len(df)-duplicate_samples)
    print('Accuracy : ',true_samples/(len(df)-duplicate_samples))
    return true_index,false_index
```

```python
true_index,false_index = calculate_predict_accuracy(total_df)
```
